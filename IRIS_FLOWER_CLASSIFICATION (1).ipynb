# **Spliting Independent and Dependent Features**

```python
x = data.iloc[:,:4]
x
```

```python
y = data['species']
y
```

# **TRAIN-TEST SPLITTING**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=0)
```

**STANDARDIZATION**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

# **MODEL CREATION**

**1) K-NEAREST NEIGHBOR (KNN) :**

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(x_train,y_train)

y_pred1=knn_model.predict(x_test)
y_pred1
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
result = confusion_matrix(y_test,y_pred1)
print(result)
labels=[0,1,2]
cmd = ConfusionMatrixDisplay(result,display_labels=labels)
```

```python
cmd.plot()
```

```python
from sklearn.metrics import accuracy_score,classification_report

print('Accuracy: ',accuracy_score(y_test,y_pred1)*100,'\n')
print(classification_report(y_test,y_pred1))
```

```python
training_score = knn_model.score(x_train,y_train)
training_score
```

**2) SUPPORT VECTOR MACHINE ALGORITHM :**


```python
from sklearn.svm import SVC

sv_model=SVC()
sv_model.fit(x_train,y_train)
y_pred2=sv_model.predict(x_test)
y_pred2
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
result = confusion_matrix(y_test,y_pred2)
print(result)
labels=[0,1,2]
cmd=ConfusionMatrixDisplay(result,display_labels=labels)
```

```python
cmd.plot()
```

```python
from sklearn.metrics import accuracy_score,classification_report

print('Accuracy: ',accuracy_score(y_test,y_pred2)*100,'\n')
print(classification_report(y_test,y_pred2))
```

```python
training_score = sv_model.score(x_train,y_train)
training_score
```

**3) LOGISTIC REGRESSION :**

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()
lr_model.fit(x_train,y_train)
y_pred3 = lr_model.predict(x_test)
y_pred3
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
result=confusion_matrix(y_test,y_pred3)
print(result)
labels=[0,1,2]
cmd=ConfusionMatrixDisplay(result,display_labels=labels)
```

```python
cmd.plot()
```

```python
from sklearn.metrics import accuracy_score,classification_report

print('Accuracy: ',accuracy_score(y_test,y_pred3)*100,'\n')
print(classification_report(y_test,y_pred3))
```

```python
training_score = lr_model.score(x_train,y_train)
training_score
```

**4) DECISION TREE :**

```python
from sklearn.tree import DecisionTreeClassifier

tree_model=DecisionTreeClassifier(max_depth = 5,criterion = 'entropy')
tree_model.fit(x_train,y_train)
y_pred4 = tree_model.predict(x_test)
y_pred4
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
result=confusion_matrix(y_test,y_pred4)
print(result)
labels=[0,1,2]
cmd=ConfusionMatrixDisplay(result,display_labels=labels)
```

```python
cmd.plot()
```

```python
from sklearn.metrics import accuracy_score,classification_report

print('Accuracy: ',accuracy_score(y_test,y_pred4)*100,'\n')
print(classification_report(y_test,y_pred4))
```

```python
training_score = tree_model.score(x_train,y_train)
training_score
```
